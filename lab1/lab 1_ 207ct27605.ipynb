```python
# @title Bai 5
import pandas as pd


file_path = 'dulieuxettuyendaihoc.csv'
data = pd.read_csv(file_path)


missing_count = data['T1'].isna().sum()
total_count = len(data)
missing_percentage = (missing_count / total_count) * 100

frequency_table = pd.DataFrame({
    'Trạng thái': ['Có dữ liệu', 'Thiếu dữ liệu'],
    'Số lượng': [total_count - missing_count, missing_count],
    'Tần suất (%)': [100 - missing_percentage, missing_percentage]
})

print(frequency_table)


mean_value = data['T1'].mean()
data['T1'].fillna(mean_value, inplace=True)


output_file_path = 'du_lieu_hieu_chinh.csv'
data.to_csv(output_file_path, index=False)
print(f'Dữ liệu đã được lưu vào file: {output_file_path}')
```

```python
# @title Bai 6
import pandas as pd

file_path = 'dulieuxettuyendaihoc.csv'
data = pd.read_csv(file_path)


score_columns = ['T1', 'T2', 'T3']


frequency_tables = {}
for column in score_columns:
    missing_count = data[column].isna().sum()
    total_count = len(data)
    missing_percentage = (missing_count / total_count) * 100

    frequency_table = pd.DataFrame({
        'Trạng thái': ['Có dữ liệu', 'Thiếu dữ liệu'],
        'Số lượng': [total_count - missing_count, missing_count],
        'Tần suất (%)': [100 - missing_percentage, missing_percentage]
    })

    frequency_tables[column] = frequency_table


for column, freq_table in frequency_tables.items():
    print(f'Tần số và tần suất cho biến {column}:')
    print(freq_table)
    print()


for column in score_columns:
    mean_value = data[column].mean()
    data[column].fillna(mean_value, inplace=True)


output_file_path = 'du_lieu_hieu_chinh.csv'
data.to_csv(output_file_path, index=False)
print(f'Dữ liệu đã được lưu vào file: {output_file_path}')
```

```python
# @title Bai 8
import pandas as pd


file_path = 'dulieuxettuyendaihoc.csv'
data = pd.read_csv(file_path)


data['TBM1'] = (data['T1']*2 + data['L1'] + data['H1'] + data['S1'] + data['V1']*2 + data['X1'] + data['D1'] + data['N1']) / 10


def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8.0:
        return 'K'
    elif tbm < 9.0:
        return 'G'
    else:
        return 'XS'


data['XL1'] = data['TBM1'].apply(xep_loai)



output_file_path = 'du_lieu_hieu_chinh.csv'
data.to_csv(output_file_path, index=False)
print(f'Dữ liệu đã được lưu vào file: {output_file_path}')
```

```python
# @title Bai 9
import pandas as pd


file_path = 'dulieuxettuyendaihoc.csv'
data = pd.read_csv(file_path)


data['TBM1'] = (data['T1']*2 + data['L1'] + data['H1'] + data['S1'] + data['V1']*2 + data['X1'] + data['D1'] + data['N1']) / 10



def min_max_normalization(tbm, min_vn=0, max_vn=10, min_us=0, max_us=4):
    return (tbm - min_vn) / (max_vn - min_vn) * (max_us - min_us) + min_us

# Tạo các biến US_TBM1, US_TBM2, US_TBM3
data['US_TBM1'] = data['TBM1'].apply(min_max_normalization)




output_file_path = 'du_lieu_hieu_chinh.csv'
data.to_csv(output_file_path, index=False)
print(f'Dữ liệu đã được lưu vào file: {output_file_path}')
```

```python
# @title Bai 10

import pandas as pd


file_path = 'dulieuxettuyendaihoc.csv'
data = pd.read_csv(file_path)


def ket_qua_xet_tuyen(row, khoi):
    if khoi == 'A' or khoi == 'A1':
        score = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif khoi == 'B':
        score = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        score = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return 1 if score >= 5.0 else 0

data['KQXT'] = data.apply(lambda row: ket_qua_xet_tuyen(row, row['DH1']), axis=1)


output_file_path = 'du_lieu_hieu_chinh.csv'
data.to_csv(output_file_path, index=False)
print(f'Dữ liệu đã được lưu vào file: {output_file_path}')
```
