```python
# @title câu 14
def classify_age(age):
    if age < 35:
        return 'tre'
    elif age < 45:
        return 'trung nien'
    else:
        return 'gia'

# Tạo cột 'age_group' dựa trên cột 'Age'
df['Nhom tuoi'] = df['Age'].apply(lambda x: classify_age(x))

print(df[['Name', 'Age', 'Nhom tuoi']])
```
